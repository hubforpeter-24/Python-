### Feature Engineering

```python
df["Revenue"] = df["Price"] * df["Installs"]
```

```python
df["Size"] = df["Size"].fillna(df['Size'].median())
```

```python
df['Size'].isnull().sum()
```

```python
# Sizing the Categories: Bin app sizes into Small, Medium & Large

size_bins = [0, 10, 50,100, float('inf')]
size_lables = ["Small","Medium", "Large", "Extra Large"]
df["Size_Category"] = pd.cut(df['Size'],bins=size_bins, labels=size_lables, right=False)
```

```python
df.head()
```

```python
df['Installs'].unique()
```

```python
df.drop(['Popularity'],inplace=True,axis=1)
```

```python
df.head()
```

```python
# Defining Popularity for Installs

popularity_bins = [0, 1e4, 1e6, 1e7, 1e9]
popularity_labels = ["Low", "Medium", "High", "Extra High"]
df['Popularity'] = pd.cut(df["Installs"], bins=popularity_bins, labels=popularity_labels, right=False)
```

```python
df['Popularity'].unique()
```

```python
df['Popularity'].isnull().sum()
```

```python
# Last Updated Year and Month
```
